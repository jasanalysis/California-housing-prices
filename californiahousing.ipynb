```python
housing_na = housing.dropna(subset = ["total_bedrooms"])
housing_na.shape
```

```python
dummies = pd.get_dummies(housing_na.ocean_proximity)
housing_na_dummies = pd.concat([housing_na,dummies], axis = 'columns')
housing_na_dummies
```

```python
housing_clean = housing_na_dummies.drop(["ocean_proximity","ISLAND"], axis = 'columns')
housing_clean.head()
```

```python
from sklearn.model_selection import train_test_split

x = housing_clean.drop(['median_house_value'], axis= 1)
y = housing_clean['median_house_value']
```

```python
# Splitting into train and test group

X_train,X_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 1984)
```

```python
from sklearn.linear_model import LinearRegression
OLS = LinearRegression()
OLS.fit(X_train,y_train)
```

```python
# Print the coefficients and intercept of the OLS line

print(OLS.intercept_)
```

```python
print("The intercept is" + str(OLS.intercept_))
print("The coefficients are" + str(OLS.coef_))
```

```python
# Finding the predicted value

y_pred = OLS.predict(X_test)
performance = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred, 'error' : y_test - y_pred})
performance.head()

```

```python
performance.reset_index(drop = True, inplace = True)
performance.reset_index(inplace = True)
performance.head()
```

```python
fig = plt.figure(figsize = (10,5))

plt.bar('index','error',data = performance, color = 'black',width = 0.3)
plt.xlabel('Observations')
plt.ylabel('Residuals')
plt.show()
```
